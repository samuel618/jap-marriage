# Data Cleaning

```python
clean_df = selected_df.copy()
# Sort by year
clean_df = clean_df.sort_values("year").reset_index(drop=True)

# Remove duplicate rows
clean_df = clean_df.drop_duplicates()

print("Shape after cleaning duplicates:", clean_df.shape)
clean_df.head()
```

```python
# Drop 1965 to keep a continuous yearly sequence from 1970 onward
clean_df = clean_df[clean_df["year"] != 1965].reset_index(drop=True)

# Convert all columns except year to numeric
for col in clean_df.columns:
    if col != "year":
        clean_df[col] = pd.to_numeric(clean_df[col], errors="coerce")

# Fill missing values with 0
clean_df = clean_df.fillna(0)

# Convert all columns to int
clean_df = clean_df.astype(int)

clean_df.info()
```

```python
# Columns where husband is Japanese and wife is foreign
husb_foreign_cols = [col for col in clean_df.columns if col.startswith("husb_jpn_wife_")]

# Columns where wife is Japanese and husband is foreign
wife_foreign_cols = [col for col in clean_df.columns if col.startswith("wife_jpn_husb_")]

print("Husband Japanese + foreign wife columns:")
print(husb_foreign_cols)

print("\nWife Japanese + foreign husband columns:")
print(wife_foreign_cols)
```

```python
# Combine all foreign-country categories
clean_df["total_husb_jpn_foreign_wife"] = clean_df[husb_foreign_cols].sum(axis=1, skipna=True)
clean_df["total_wife_jpn_foreign_husband"] = clean_df[wife_foreign_cols].sum(axis=1, skipna=True)

# Total marriages involving a foreign spouse
clean_df["total_foreign_spouse"] = (
    clean_df["total_husb_jpn_foreign_wife"] +
    clean_df["total_wife_jpn_foreign_husband"]
)

# Total marriages
clean_df["total_marriages"] = (
    clean_df["both_japanese"] +
    clean_df["total_foreign_spouse"]
)
clean_df.head()
```

# Data Preparation

```python
jap_copy_df = clean_df[
    [
        "year",
        "both_japanese",
        "total_husb_jpn_foreign_wife",
        "total_wife_jpn_foreign_husband",
        "total_foreign_spouse",
        "total_marriages"
    ]
].copy()

jap_copy_df.head()
```

```python
jap_df = jap_copy_df.melt(
    id_vars="year",
    var_name="type",
    value_name="amount"
)

jap_df.head()
```

# Model Training

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
# Make sure required columns are valid
jap_df["year"] = pd.to_numeric(jap_df["year"], errors="coerce")
jap_df["amount"] = pd.to_numeric(jap_df["amount"], errors="coerce")

# Drop rows with missing values in required columns
jap_df = jap_df.dropna(subset=["year", "type", "amount"]).copy()

# Sort data
jap_df = jap_df.sort_values(["type", "year"]).reset_index(drop=True)

jap_df.head()
```

```python
train_cutoff = 2020
future_year = 2025
degree = 2

poly_results = []
poly_predictions = []

categories = jap_df["type"].unique()

for category in categories:
    subset = jap_df[jap_df["type"] == category].copy()
    
    train_df = subset[subset["year"] <= train_cutoff].copy()
    test_df = subset[subset["year"] > train_cutoff].copy()
    
    if train_df.empty or test_df.empty:
        print(f"Skipping {category}: not enough train/test data.")
        continue
    
    X_train = train_df[["year"]]
    y_train = train_df["amount"]
    
    X_test = test_df[["year"]]
    y_test = test_df["amount"]
    
    # Polynomial Regression model
    model = Pipeline([
        ("poly", PolynomialFeatures(degree=degree, include_bias=False)),
        ("linear", LinearRegression())
    ])
    
    model.fit(X_train, y_train)
    
    # Predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    pred_2025 = model.predict(pd.DataFrame({"year": [future_year]}))[0]
    
    # Metrics
    train_r2 = r2_score(y_train, y_train_pred)
    test_r2 = r2_score(y_test, y_test_pred)
    
    train_mae = mean_absolute_error(y_train, y_train_pred)
    test_mae = mean_absolute_error(y_test, y_test_pred)
    
    # Save summary
    poly_results.append({
        "type": category,
        "degree": degree,
        "train_start_year": int(train_df["year"].min()),
        "train_end_year": int(train_df["year"].max()),
        "test_start_year": int(test_df["year"].min()),
        "test_end_year": int(test_df["year"].max()),
        "train_r2": train_r2,
        "test_r2": test_r2,
        "train_mae": train_mae,
        "test_mae": test_mae,
        "prediction_2025": pred_2025
    })
    
    # Save detailed predictions
    train_temp = train_df.copy()
    train_temp["set"] = "train"
    train_temp["predicted_amount"] = y_train_pred
    
    test_temp = test_df.copy()
    test_temp["set"] = "test"
    test_temp["predicted_amount"] = y_test_pred
    
    combined_temp = pd.concat([train_temp, test_temp], ignore_index=True)
    poly_predictions.append(combined_temp)
```

```python
results_df = pd.DataFrame(poly_results)
results_df
```

```python
predictions_df = pd.concat(poly_predictions, ignore_index=True)
predictions_df.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

label_map = {
    "both_japanese": "Both Japanese",
    "total_husb_jpn_foreign_wife": "Japanese Husband + Foreign Wife",
    "total_wife_jpn_foreign_husband": "Japanese Wife + Foreign Husband",
    "total_foreign_spouse": "Total Foreign Spouse",
    "total_marriages": "Total Marriages"
}

degree = 3
forecast_years = np.arange(2025, 2031)  # 2025 to 2030

for category in results_df["type"]:
    subset = jap_df[jap_df["type"] == category].copy().sort_values("year")
    
    X = subset[["year"]]
    y = subset["amount"]
    
    # Train polynomial model on full historical data
    model = Pipeline([
        ("poly", PolynomialFeatures(degree=degree, include_bias=False)),
        ("linear", LinearRegression())
    ])
    model.fit(X, y)
    
    # Smooth curve for historical period
    historical_years = np.linspace(subset["year"].min(), subset["year"].max(), 300)
    historical_years_df = pd.DataFrame({"year": historical_years})
    historical_pred = model.predict(historical_years_df)
    
    # Forecast years 2025-2030
    forecast_df = pd.DataFrame({"year": forecast_years})
    forecast_pred = model.predict(forecast_df)
    
    nice_name = label_map.get(category, category)
    
    plt.figure(figsize=(10, 5))
    
    # Historical actual data
    plt.scatter(subset["year"], subset["amount"], label="Historical Data")
    
    # Historical fitted curve
    plt.plot(historical_years, historical_pred, label="Polynomial Regression Curve")
    
    # Forecast line
    plt.plot(forecast_years, forecast_pred, linestyle="--", label="Forecast (2025-2030)")
    
    # Forecast points
    plt.scatter(forecast_years, forecast_pred, s=80, label="Forecast Points")
    
    plt.title(f"{nice_name}: Historical Data and Forecast to 2030")
    plt.xlabel("Year")
    plt.ylabel("Amount")
    plt.legend()
    plt.grid(True)
    plt.show()
```

# Metrics

```python
eval_results = []
train_cutoff = 2020
degree = 3

categories = jap_df["type"].unique()

for category in categories:
    subset = jap_df[jap_df["type"] == category].copy()
    
    train_df = subset[subset["year"] <= train_cutoff].copy()
    test_df = subset[subset["year"] > train_cutoff].copy()
    
    if train_df.empty or test_df.empty:
        print(f"Skipping {category}: not enough train/test data.")
        continue
    
    X_train = train_df[["year"]]
    y_train = train_df["amount"]
    
    X_test = test_df[["year"]]
    y_test = test_df["amount"]
    
    # Polynomial regression model
    model = Pipeline([
        ("poly", PolynomialFeatures(degree=degree, include_bias=False)),
        ("linear", LinearRegression())
    ])
    
    model.fit(X_train, y_train)
    
    # Predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    # Train metrics
    train_mae = mean_absolute_error(y_train, y_train_pred)
    train_mse = mean_squared_error(y_train, y_train_pred)
    train_r2 = r2_score(y_train, y_train_pred)
    
    # Test metrics
    test_mae = mean_absolute_error(y_test, y_test_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)
    test_r2 = r2_score(y_test, y_test_pred)
    
    eval_results.append({
        "type": category,
        "train_mae": train_mae,
        "train_mse": train_mse,
        "train_r2": train_r2,
        "test_mae": test_mae,
        "test_mse": test_mse,
        "test_r2": test_r2
    })
```

```python
evaluation_df = pd.DataFrame(eval_results)
evaluation_df
```
